### **FUNCIÓN PARA ELIMINAR OUTLIERS**

```python
def eliminar_outliers(df, columns):
    df_copia=df.copy()
    for column in columns:
        Q1 = df_copia[column].quantile(0.25) #Calcula el primer cuartil (Q1), que es el valor por debajo del cual se encuentra el 25% de los datos de la columna actual.
        Q3 = df_copia[column].quantile(0.75) #Calcula el tercer cuartil (Q3), que es el valor por debajo del cual se encuentra el 75% de los datos de la columna actual.
        IQR = Q3 - Q1 #Calcula el rango intercuartílico (IQR), que es la diferencia entre el tercer y el primer cuartil. El IQR representa la dispersión de la parte central del conjunto de datos
        lim_inf = Q1 - 1.5 * IQR #Calcula el límite inferior para detectar outliers. Los valores por debajo de este límite se considerarán outliers.
        lim_sup = Q3 + 1.5 * IQR #Calcula el límite superior para detectar outliers. Los valores por encima de este límite se considerarán outliers.
        filtro_outliers = (df_copia[column] < lim_inf) | (df_copia[column] > lim_sup) #Creamos la máscara, el filtro que se va a aplicar al df original para eliminar outliers
        df_outliers_elim=df_copia[~filtro_outliers]
    return df_outliers_elim
```

### **VARIABLES MÁS IMPORTANTES**

*El arbol de decisión nos va a permitir saber que variable es más importante, asignando pesos a las variables dependiendo de su influencia*

```python
# Importar el algoritmo de árboles de decisión y las métricas correspondientes
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Generar la lista de features y la variable target
target = 'Revenue'
features = [x for x in df_online_shoppers_normalizado.columns if x!=target] #Creamos las características seleccionando todas las columnas excepto la columna de la variable objetivo "Revenue".

# Asignar el algortimo e indicar la profundidad máxima del árbol (con un número rotundamente grande para sobreajustar)
arbol_importancia = DecisionTreeRegressor(max_depth=len(features)+10, random_state=100)

# Entrenar un árbol con todo el conjunto de datos
arbol_importancia.fit(X=df_online_shoppers_normalizado[features], y=df_online_shoppers_normalizado[target])

# Comprobamos que se obtiene un R^2 muy alto. Lo desesable es que sea 1.
y_pred_arbol = arbol_importancia.predict(X=df_online_shoppers_normalizado[features])

# Métricas para evaluar la calidad del modelo
print('Mean Absolute Error:', mean_absolute_error(df_online_shoppers_normalizado[target], y_pred_arbol))
print('Mean Absolute Percentage Error:', mean_absolute_percentage_error(df_online_shoppers_normalizado[target], y_pred_arbol)*100)
print('Mean Squared Error:', mean_squared_error(df_online_shoppers_normalizado[target], y_pred_arbol))
print('Root Mean Squared Error:', np.sqrt(mean_squared_error(df_online_shoppers_normalizado[target], y_pred_arbol)))
print('R^2 coefficient of determination:', r2_score(df_online_shoppers_normalizado[target], y_pred_arbol))
```

```python
# Creamos un DataFrame con los datos de importancia
importancia_variables = pd.DataFrame(data=arbol_importancia.feature_importances_, index=features, columns=["Importancia"])

# Ordenamos los datos
importancia_variables.sort_values(by=importancia_variables.columns[0], ascending=False, inplace=True)
importancia_variables
```

### **KMEANS**

*El kmeans particiona un conjunto de datos en grupos o clusters, donde cada punto de datos pertenece al cluster cuyo centroide es más cercano. Intenta minimizar la suma de las distancias cuadradas de cada punto al centroide de su cluster asignado*

#### **GRID SEARCH**

*El grid search nos va a permitir poder buscar el modelo óptimo, probando varios modelos, utilizando el rango de parámetros que le pasemos*

```python
#Importaciones necesarias
from sklearn.model_selection import KFold
from sklearn.metrics import make_scorer
from sklearn.model_selection import GridSearchCV
```

```python
# Definimos la manera de trozear dataset de train para crear los subconjuntos que se utilizan en la validación cruzada (Cross-Validation)
folds_cv_dtc_onshop= KFold(n_splits=10, shuffle=True, random_state=111) #vamos a hacer en 10, cada vez sacas uno, con los nueves restantes entrenas el modelo y con el sobrante lo testeas.

# Declaramos el tipo de regresor (modelo) y los rangos de hiperparámetros a considerar en la búsqueda
modelo_dtc_grid_onshop = DecisionTreeClassifier(random_state=100)
parametros_grid_onshop = {"max_depth":np.arange(2, 50), "max_features":np.arange(1, 17)} #de golpe nos permite estudiar diferentes valores para nuestro hiperparametro.
roc_auc_scorer_dtc_onshop = make_scorer(roc_auc_score)

# Definir el Grid Search y realizar la búsqueda con los datos de entrenamiento
#un gridSearch, dice que modelo quieres que haga, mi modelo es un DecisionTreeClassifier , q parametros queires q considere? (parametros = {"max_depth":list(range(2,20))} ) 
# y en q quieres q me base?, pues en el roc_auc_score, por eso ponemos abajo scoring='roc_auc_score',Establecer el roc_auc_score como la función de "scoring" a la hora de puntuar los modelos

modelo_grid_dtc_onshop = GridSearchCV(estimator=modelo_dtc_grid_onshop,
                          param_grid=parametros_grid_onshop,
                          scoring='roc_auc',
                          cv=folds_cv_dtc_onshop,
                          n_jobs=-1)  
modelo_grid_dtc_onshop.fit(X_train_onshop, y_train_onshop)
```

#### **CURVA DE APRENDIZAJE**

##### **DEFINICIÓN FUNCIÓN CURVA DE APRENDIZAJE**

```python
def curva_aprendizaje_dtc_onshop(profundidad: int=None):
    """Función para sacar la gráfica de Learning Curve a partir de la profundidad del árbol"""

    # Se toma el valor entero y si no se indica nada por defecto se pasa el None como la profundidad máxima
    if profundidad:
        hiper_parametro = int(profundidad)
    else:
        hiper_parametro = None

    # Generar la estructura del árbol
    modelo_dtc_onshop = DecisionTreeClassifier(max_depth=hiper_parametro, random_state=100)
    
    # Crear un listado de los tamaños de subconjuntos de datos de entrenamiento
    num_samples = np.linspace(100,X_train_onshop.shape[0],100).astype(int)

    # Generamos previamente los vectores necesarios para ir calculando y guardando el rendimiento de los árboles
    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        # Generamos un árbol para cada subconjunto de datos de entrenamiento y lo ajustamos  
        modelo_dtc_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])

        # Calculamos el área bajo la curva de funcionamiento del receptor sobre datos de train y de test
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]], modelo_dtc_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_dtc_onshop.predict(X_test_onshop))

    print("La diferencia final del rendimiento del modelo entre training y test es= ", train_score[-1]-test_score[-1])

    # Graficamos el rendimiento de training versus de test
    plt.plot(num_samples, test_score, label = 'Test Accuracy')
    plt.plot(num_samples, train_score, label = 'Train Accuracy')
    plt.title('Curva de aprendizaje (Learning Curve) para la profundidad = %s' % hiper_parametro)
    plt.xlabel('Tamaño de entrenamiento (Training size)')
    plt.ylabel('Accuracy')
    plt.legend(loc = 'lower right')
    plt.show()    
```

```python
curva_aprendizaje_dtc_onshop(5)
```

**OBSERVACIONES:** *Veo una buena curva de aprendizaje ya que no hay un gran gap entre train y test, parece que hay continuidad y que en algún momento se casarán y se mantienen en unos números buenos*

#### **CURVA DE APRENDIZAJE**

##### **DEFINICION FUNCIÓN CURVA DE APRENDIZAJE**

```python
def curva_aprendizaje_rfc_onshop(n_arboles: int,
                               profundidad: int):
    """Función para sacar la gráfica de Learning Curve a partir de los hiperparámetros del bosque"""

    # Generar la estructura del modelo
    modelo_rfc_onshop = RandomForestClassifier(n_estimators=int(n_arboles),
                                          max_depth=int(profundidad),
                                          oob_score=True,
                                          n_jobs=-1,
                                          random_state=100)
    
    # Crear un listado de los tamaños de subconjuntos de datos de entrenamiento
    num_samples = np.linspace(100,X_train_onshop.shape[0]).astype(int)

    # Generamos previamente los vectores necesarios para ir calculando y guardando el rendimiento de los modelos
    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)
    oob_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        # Generamos un modelo para cada subconjunto de datos de entrenamiento y lo ajustamos  
        modelo_rfc_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])

        # Calculamos el R_cuadrado sobre datos de train y de test
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]],modelo_rfc_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_rfc_onshop.predict(X_test_onshop))
        oob_score[i] = modelo_rfc_onshop.oob_score_

    print("La diferencia accuracy_score entre training y OOB es= ", train_score[-1]-oob_score[-1])

    # Graficamos el R_cuadrado de training versus de test
    fig = plt.figure(figsize=(20,7))
    plt.plot(num_samples, test_score, label = 'Test Accuracy')
    plt.plot(num_samples, train_score, label = 'Train Accuracy')
    plt.plot(num_samples, oob_score, label = 'OOB R^2')
    plt.title(f'Curva de aprendizaje (Learning Curve) para modelo Random Forest Classifier con profundidad {profundidad} y nº de árboles {n_arboles}')
    plt.xlabel('Tamaño de entrenamiento (Training size)')
    plt.ylabel('Accuracy')
    plt.legend(loc = 'lower right')
    plt.show()    
```

**OBSERVACIONES:** *En esta curva de aprendizaje puedo ver que hay un gap bastante notable y que aunque el test tiene un poco de continuidad, veo que hay overfitting ya que el train es muy alto y se mantiene, mientras que el test aunque no es malo del todo es mucho más bajo, por lo que no generaliza adecuadamente, teniendo un sobreajuste*

```python
def curva_aprendizaje_xgb_onshop(n_arboles: int,
                               profundidad: int
                               ):
    """Función para sacar la gráfica de Learning Curve a partir de los hiperparámetros del bosque"""

    # Generar la estructura del modelo
    modelo_xgb_onshop = XGBClassifier(n_estimators=int(n_arboles),
                                          max_depth=int(profundidad),
                                          n_jobs=-1,
                                          random_state=100)
    
    # Crear un listado de los tamaños de subconjuntos de datos de entrenamiento
    num_samples = np.linspace(100,X_train_onshop.shape[0]).astype(int)

    # Generamos previamente los vectores necesarios para ir calculando y guardando el rendimiento de los modelos
    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        # Generamos un modelo para cada subconjunto de datos de entrenamiento y lo ajustamos  
        modelo_xgb_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])

        # Calculamos el R_cuadrado sobre datos de train y de test
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]],modelo_xgb_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_xgb_onshop.predict(X_test_onshop))

    print("La diferencia accuracy_score entre training y test es= ", train_score[-1]-test_score[-1])

    # Graficamos el R_cuadrado de training versus de test
    fig = plt.figure(figsize=(20,7))
    plt.plot(num_samples, test_score, label = 'Test Accuracy')
    plt.plot(num_samples, train_score, label = 'Train Accuracy')
    plt.title('Curva de aprendizaje (Learning Curve) para la profundidad = %s' % profundidad)
    plt.xlabel('Tamaño de entrenamiento (Training size)')
    plt.ylabel('Accuracy')
    plt.legend(loc = 'lower right')
    plt.show()    
```

## **COMPARACIÓN POR CURVA DE APRENDIZAJE**

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier

def curva_aprendizaje_xgb_onshop(n_arboles: int, profundidad: int, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop):
    modelo_xgb_onshop = XGBClassifier(n_estimators=n_arboles, max_depth=profundidad, n_jobs=-1, random_state=100)
    
    num_samples = np.linspace(100, X_train_onshop.shape[0], 10).astype(int)

    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)
    diff_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        modelo_xgb_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]], modelo_xgb_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_xgb_onshop.predict(X_test_onshop))
        diff_score[i] = train_score[i] - test_score[i]

    return num_samples, train_score, test_score, diff_score

def curva_aprendizaje_rfc_onshop(n_arboles: int, profundidad: int, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop):
    modelo_rfc_onshop = RandomForestClassifier(n_estimators=n_arboles, max_depth=profundidad, oob_score=True, n_jobs=-1, random_state=100)
    
    num_samples = np.linspace(100, X_train_onshop.shape[0], 10).astype(int)

    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)
    oob_score = np.zeros(num_samples.size)
    diff_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        modelo_rfc_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]], modelo_rfc_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_rfc_onshop.predict(X_test_onshop))
        oob_score[i] = modelo_rfc_onshop.oob_score_
        diff_score[i] = train_score[i] - oob_score[i]

    return num_samples, train_score, test_score, oob_score, diff_score

def curva_aprendizaje_dtc_onshop(profundidad: int, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop):
    hiper_parametro = profundidad if profundidad else None
    modelo_dtc_onshop = DecisionTreeClassifier(max_depth=hiper_parametro, random_state=100)
    
    num_samples = np.linspace(100, X_train_onshop.shape[0], 10).astype(int)

    train_score = np.zeros(num_samples.size) 
    test_score = np.zeros(num_samples.size)
    diff_score = np.zeros(num_samples.size)

    for i in range(num_samples.size):
        modelo_dtc_onshop.fit(X_train_onshop[:num_samples[i]], y_train_onshop[:num_samples[i]])
        train_score[i] = accuracy_score(y_train_onshop[:num_samples[i]], modelo_dtc_onshop.predict(X_train_onshop[:num_samples[i]])) 
        test_score[i] = accuracy_score(y_test_onshop, modelo_dtc_onshop.predict(X_test_onshop))
        diff_score[i] = train_score[i] - test_score[i]

    return num_samples, train_score, test_score, diff_score

# Función para graficar las curvas de aprendizaje juntas
def plot_learning_curves(num_samples, train_scores, test_scores, oob_scores=None, diff_scores=None):
    fig, axes = plt.subplots(1, 3, figsize=(20, 5), sharex=True, sharey=True)

    # Graficar XGBoost
    axes[0].plot(num_samples[0], test_scores[0], label='Precisión de prueba')
    axes[0].plot(num_samples[0], train_scores[0], label='Precisión de entrenamiento')
    diff_mean_xgb = np.mean(diff_scores[0])
    axes[0].set_title('Curva de aprendizaje de XGBoost')
    axes[0].set_xlabel('Tamaño de entrenamiento')
    axes[0].set_ylabel('Precisión')
    axes[0].legend(loc='lower right')
    axes[0].text(0.5, -0.15, f'Diferencia entre Train y Test: {diff_mean_xgb:.4f}', horizontalalignment='center', verticalalignment='center', transform=axes[0].transAxes)

    # Graficar Random Forest
    axes[1].plot(num_samples[1], test_scores[1], label='Precisión de prueba')
    axes[1].plot(num_samples[1], train_scores[1], label='Precisión de entrenamiento')
    if oob_scores is not None:
        axes[1].plot(num_samples[1], oob_scores[1], label='Precisión OOB')
        if diff_scores is not None:
            diff_mean_rfc = np.mean(diff_scores[1])
            axes[1].text(0.5, -0.15, f'Diferencia entre Train y OOB: {diff_mean_rfc:.4f}', horizontalalignment='center', verticalalignment='center', transform=axes[1].transAxes)
    axes[1].set_title('Curva de aprendizaje de Random Forest')
    axes[1].set_xlabel('Tamaño de entrenamiento')
    axes[1].legend(loc='lower right')

    # Graficar Decision Tree
    axes[2].plot(num_samples[2], test_scores[2], label='Precisión de prueba')
    axes[2].plot(num_samples[2], train_scores[2], label='Precisión de entrenamiento')
    diff_mean_dtc = np.mean(diff_scores[2])
    axes[2].set_title('Curva de aprendizaje de Decision Tree')
    axes[2].set_xlabel('Tamaño de entrenamiento')
    axes[2].legend(loc='lower right')
    axes[2].text(0.5, -0.15, f'Diferencia entre Train y Test: {diff_mean_dtc:.4f}', horizontalalignment='center', verticalalignment='center', transform=axes[2].transAxes)

    plt.tight_layout()
    plt.show()

# Ejemplo de uso:
# Suponiendo que tienes X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop definidos anteriormente
n_arboles = 100
profundidad = 5

num_samples_xgb, train_scores_xgb, test_scores_xgb, diff_scores_xgb = curva_aprendizaje_xgb_onshop(29, 6, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop)
num_samples_rfc, train_scores_rfc, test_scores_rfc, oob_scores_rfc, diff_scores_rfc = curva_aprendizaje_rfc_onshop(200, 26, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop)
num_samples_dtc, train_scores_dtc, test_scores_dtc, diff_scores_dtc = curva_aprendizaje_dtc_onshop(5, X_train_onshop, y_train_onshop, X_test_onshop, y_test_onshop)

plot_learning_curves([num_samples_xgb, num_samples_rfc, num_samples_dtc],
                     [train_scores_xgb, train_scores_rfc, train_scores_dtc],
                     [test_scores_xgb, test_scores_rfc, test_scores_dtc],
                     [None, oob_scores_rfc, None],
                     [diff_scores_xgb, diff_scores_rfc, diff_scores_dtc])
```
